```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
from IPython.display import Image  #To display image directly in our workspace
```

```python
# Read image on gray scale

checkerboard = cv2.imread('checkerboard.jpg', 0)
plt.imshow(checkerboard, cmap='gray')
print(checkerboard)
```

```python
# To access individual pixel of the first black box
print(checkerboard[0,0])

# Print the first white pixel
print(checkerboard[0,200])
```

#### Modifying Image Pixels

```python
checkerboard_new = checkerboard.copy()
# checkerboard_new[30, 30]= 200
# checkerboard_new[30, 100]= 200
# checkerboard_new[100, 30]= 200
# checkerboard_new[100, 100]= 200

checkerboard_new[30:100, 30:100] = 255

plt.imshow(checkerboard_new, cmap='gray')
print(checkerboard_new)
```

#### Cropping images

```python
MyPic_BGR = cv2.imread('MyPic.jpg', cv2.IMREAD_COLOR)
MyPic_RGB = MyPic_BGR[:,:,::-1]
plt.imshow(MyPic_RGB)
```

```python
# To crop the image
MyPic_Crop = MyPic_RGB[60:450, 200:550]
plt.imshow(MyPic_Crop)
```

#### Resizing images

```python
# Resizing by factors

resized_MyPic_Crop = cv2.resize(MyPic_Crop, None, fx=2, fy=2)
plt.imshow(resized_MyPic_Crop)
```

```python
# Specifying the exact size of the output image

new_width = 100
new_height = 200
dim = (new_width, new_height)

Resized_MyPic_Crop = cv2.resize(MyPic_Crop, dsize = dim, interpolation = cv2.INTER_AREA)
plt.imshow(Resized_MyPic_Crop)
```

```python
# Resizing while maintaing the aspect ratio

new_width = 100
aspect_ratio = new_width/MyPic_Crop.shape[1]
new_height = int(MyPic_Crop.shape[0] * aspect_ratio)
dim = (new_width, new_height)

# Resize image
resized_aspect = cv2.resize(MyPic_Crop, dsize = dim, interpolation =cv2.INTER_AREA)
plt.imshow(resized_aspect)
```

#### Flipping and Rotating Images

```python
# Flipping Images

my_Image_flipped_Horizontally = cv2.flip(MyPic_RGB, 1)
my_Image_flipped_Vertically = cv2.flip(MyPic_RGB, 0)
my_Image_flipped_Both = cv2.flip(MyPic_RGB, -1)

# Let see the images
plt.figure(figsize = [18, 5])
plt.subplot(141); plt.title('Original'); plt.imshow( MyPic_RGB)
plt.subplot(142); plt.title('Horizontal Flip'); plt.imshow(my_Image_flipped_Horizontally)
plt.subplot(143); plt.title('Vertically Flip'); plt.imshow(my_Image_flipped_Vertically)
plt.subplot(144); plt.title('Both Flipped'); plt.imshow(my_Image_flipped_Both)
```

```python
# Rotating Images

my_Image_rotate_90 = cv2.rotate(MyPic_RGB, 0)
my_Image_rotate_180 = cv2.rotate(MyPic_RGB, 1)
my_Image_rotate_270 = cv2.rotate(MyPic_RGB, 2)

# Let see the images
plt.figure(figsize = [18, 5])
plt.subplot(141); plt.title('Original'); plt.imshow( MyPic_RGB)
plt.subplot(142); plt.title('90 degree Rotation'); plt.imshow(my_Image_rotate_90)
plt.subplot(143); plt.title('180 degree Rotation'); plt.imshow(my_Image_rotate_180)
plt.subplot(144); plt.title('270 degree Rotation'); plt.imshow(my_Image_rotate_270)
```
