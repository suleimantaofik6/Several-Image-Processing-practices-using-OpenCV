```python
# To implement the Canny edge detections, we need 5 steps:

# 1. Noise reduction by applying gausian filter to smooth the image and remove noise
# 2. Gradient Calculation: Find the intensity gradient of the image
# 3. Non-Maximum suppression to get rid of spourious response to edge detection
# 4. Double threshold to determine potential edges
# 5. Edge Tracking by hysteresis i.e. to finalize the detection of the edges by supressing other edges that are weak or not connected to strong edges.
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('canny.png', 0)
canny = cv2.Canny(img, 100, 200)

titles = ['image', 'canny']
images = [img, canny]

for i in range(2):
    plt.figure(figsize = [15, 5]), plt.subplot(1,2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```
