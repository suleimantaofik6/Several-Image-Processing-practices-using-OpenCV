#### Changing images to different colour spaces

```python
# To convert from BGR TO RGB
flower_rgb = cv2.cvtColor(flower_bgr, cv2.COLOR_BGR2RGB)
plt.figure(figsize =[20, 5])
plt.subplot(141); plt.title('BGR Image'); plt.imshow(flower_bgr)
plt.subplot(142); plt.title('Original image'); plt.imshow(flower_rgb)
```

```python
# To convert to Hue Saturation and Value (HSV) colour space
flower_hsv = cv2.cvtColor(flower_bgr, cv2.COLOR_BGR2HSV)

# Split the image into the H, S, V components
h,s,v =cv2.split(flower_hsv)

# Show the new channels:
plt.figure(figsize =[20, 5])
plt.subplot(141); plt.title('H channel'); plt.imshow(h, cmap='gray')
plt.subplot(142); plt.title('S channel'); plt.imshow(s, cmap='gray')
plt.subplot(143); plt.title('V channel'); plt.imshow(v, cmap='gray')
plt.subplot(144); plt.title('Original image'); plt.imshow(flower_rgb)
```

```python
# Let modify one of the channel and compare it to our original image

h_new = h + 10
flower_merged = cv2.merge((h_new, s, v))
hflower_rgb = cv2.cvtColor(flower_merged, cv2.COLOR_HSV2RGB)

# Let see the plot
plt.figure(figsize =[20, 5])
plt.subplot(141); plt.title('Modified image'); plt.imshow(hflower_rgb)
plt.subplot(142); plt.title('Original image'); plt.imshow(flower_rgb)
```

```python
# Let save the new image using imwrite
cv2.imwrite('Flower_h10.png', hflower_rgb)
Image('flower_h10.png')
```
