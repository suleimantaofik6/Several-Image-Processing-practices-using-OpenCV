```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)
from IPython.display import Image  #To display image directly in our workspace
```

```python
Apollo = cv2.imread('Apollo.jpg', cv2.IMREAD_COLOR)

# Disply the image
plt.imshow(Apollo[:,:,::-1])
```

#### Drawing a line, circle, rectangle

```python
# Drawing a line
ApolloLine = Apollo.copy()
cv2.line(ApolloLine, (50, 40), (100, 40), (0, 255, 255), thickness =1, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(ApolloLine[:,:,::-1])
```

```python
# Add a circle

ApolloCircle = Apollo.copy()
cv2.circle(ApolloCircle, (225, 120), 20, (0, 0, 255), thickness =1, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(ApolloCircle[:,:,::-1])
```

```python
# To draw a Rectangle

ApolloRectangle = Apollo.copy()
cv2.rectangle(ApolloRectangle, (110, 20), (165, 140), (255, 0, 255), thickness =1, lineType=cv2.LINE_8);

# Display the image
plt.imshow(ApolloRectangle[:,:,::-1])
```

#### Adding Text

```python
ApolloText= Apollo.copy()
text = 'Apollo 11 Saturn V Launch, July 16, 1969'
fontScale = 0.6
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 0)
fontThickness = 1

cv2.putText(ApolloText, text, (50, 160), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA)

# Display the image
plt.imshow(ApolloText)
```
